Use the cell below to implement the requested API by means of **AVL Tree** (if among your chosen data structure).

```python
class AVLTree(AbstractSearchInterface):

    def __init__(self):
        self.root = None


    def insertElement(self, element):
        inserted = False
        #Only insert if element doesn't already exist
        if not self.searchElement(element):
            self.root = self._insert(self.root, element)
            inserted = True

        return inserted


    def searchElement(self, element):
        found = False
        current = self.root

        #Traverse through nodes knowing they're ordered
        while current:
            if element == current.value:
                found = True
                break
            elif element < current.value:
                current = current.left
            else:
                current = current.right

        return found


    def _insert(self, node, element):
        if not node:
            return AVLNode(element)

        if element < node.value:
            node.left = self._insert(node.left, element)
        else:
            node.right = self._insert(node.right, element)

        node.height = 1 + max(self._get_height(node.left), self._get_height(node.right))

        balance = self._get_balance(node)

        #Choose rotation based on balance
        if balance > 1 and element < node.left.value:
            return self._right_rotate(node)
        if balance < -1 and element > node.right.value:
            return self._left_rotate(node)
        if balance > 1 and element > node.left.value:
            node.left = self._left_rotate(node.left)
            return self._right_rotate(node)
        if balance < -1 and element < node.right.value:
            node.right = self._right_rotate(node.right)
            return self._left_rotate(node)

        return node

    def _get_height(self, node):
        if not node:
            return 0
        return node.height

    def _get_balance(self, node):
        if not node:
            return 0
        return self._get_height(node.left) - self._get_height(node.right)

    #Rotate entire tree around pivot node if balance is wrong
    def _left_rotate(self, node):
        new_root = node.right
        node.right = new_root.left
        new_root.left = node

        node.height = 1 + max(self._get_height(node.left), self._get_height(node.right))
        new_root.height = 1 + max(self._get_height(new_root.left), self._get_height(new_root.right))

        return new_root

    def _right_rotate(self, node):
        new_root = node.left
        node.left = new_root.right
        new_root.right = node

        node.height = 1 + max(self._get_height(node.left), self._get_height(node.right))
        new_root.height = 1 + max(self._get_height(new_root.left), self._get_height(new_root.right))

        return new_root
```

Use the cell below to implement the requested API by means of **LLRB BST** (if among your chosen data structure).

```python
class LLRBBST(AbstractSearchInterface):


    def insertElement(self, element):
        inserted = False
        self.root, inserted = self._insert(self.root, element)
        if self.root is not None:
            self.root.colour = BLACK
        return inserted  # Root will always be black

    def searchElement(self, element): # Traverse according to BST ordering
        found = False
        node = self.root
        while node is not None:
            if element < node.key:
                node = node.left
            elif element > node.key:
                node = node.right
            else:
                return True
        return found

    def __init__(self):
        self.root = None

    def _is_red(self, node):
        if node is None:
            return False
        return node.colour == RED

    def _rotate_left(self, node):         # Fixes right leaning red link
        child = node.right
        node.right = child.left
        child.left = node
        child.colour = node.colour
        node.colour = RED
        return child

    def _rotate_right(self, node):        # Fix two left red links consecutively
        child = node.left
        node.left = child.right
        child.right = node
        child.colour = node.colour
        node.colour = RED
        return child

    def _colour_flip(self, node):         # Split 4-node by pushing colour up to the parent
        node.colour = not node.colour
        node.left.colour = not node.left.colour
        node.right.colour = not node.right.colour

    def _insert(self, node, element):
        if node is None:
            return LLRBNode(element), True

        if element < node.key:
            node.left, inserted = self._insert(node.left, element)
        elif element > node.key:
            node.right, inserted = self._insert(node.right, element)
        else:
            return node, False

        #fix LLRB invariants on way up
        if self._is_red(node.right) and not self._is_red(node.left):
            node = self._rotate_left(node)
        if self._is_red(node.left) and self._is_red(node.left.left):
            node = self._rotate_right(node)
        if self._is_red(node.left) and self._is_red(node.right):
            self._colour_flip(node)

        return node, inserted
```

Use the cell below to implement the requested API by means of **B-Tree** (if among your chosen data structure).

```python
class BTree(AbstractSearchInterface):
    def __init__(self, t=2):
        self.t = t
        self.root = BTreeNode(leaf=True)

    def insertElement(self, element):
        inserted = False

        # Skip if already inserted
        if self.searchElement(element):
            return inserted

        root = self.root

        if len(root.keys) == 2 * self.t - 1:
            new_root = BTreeNode(leaf=False)
            new_root.children.append(self.root)
            self._split_child(new_root, 0)
            self.root = new_root

        self._insert_non_full(self.root, element)
        inserted = True

        return inserted

    def _insert_non_full(self, node, element):
        i = len(node.keys) - 1

        if node.leaf:
            # New empty key so we can start shifting right
            node.keys.append(None)
            while i >= 0 and element < node.keys[i]:
              node.keys[i+1] = node.keys[i]
              i -= 1
            node.keys[i+1] = element
        else:
            while i >= 0 and element < node.keys[i]:
                i -= 1
            i += 1

            # If that child is full, split it first
            if len(node.children[i].keys) == 2 * self.t - 1:
                self._split_child(node, i)
                # After split, decide who gets the key
                if element > node.keys[i]:
                    i += 1

            self._insert_non_full(node.children[i], element)

    def _split_child(self, parent, index):
        t = self.t
        full_node = parent.children[index]
        new_node = BTreeNode(leaf=full_node.leaf)

        median_key = full_node.keys[t - 1]

        # New node gets right half of keys
        new_node.keys = full_node.keys[t:]
        full_node.keys = full_node.keys[: t - 1]

        # If not leaf, new node also gets the right half of children
        if not full_node.leaf:
            new_node.children = full_node.children[t:]
            full_node.children = full_node.children[:t]

        parent.keys.insert(index, median_key)
        parent.children.insert(index + 1, new_node)

    def _search(self, node, element):
        i = 0
        while i < len(node.keys) and element > node.keys[i]:
            i += 1

        if i < len(node.keys) and element == node.keys[i]:
            return True

        if node.leaf:
            return False

        return self._search(node.children[i], element)


    def searchElement(self, element):
        found = False

        found = self._search(self.root, element)

        return found
```

```python
import string
import random

class TestDataGenerator():
    '''
    A class to represent a synthetic data generator.

    ...

    Attributes
    ----------

    [to be defined as part of the coursework]

    Methods
    -------

    [to be defined as part of the coursework]

    '''

    #ADD YOUR CODE HERE

    def __init__(self, element_count=100, length=5, random_length=False):
        self.element_count = element_count
        self.length = length

    def generate_random_string(self):
        letters = string.ascii_lowercase
        return ''.join(random.choice(letters) for _ in range(self.length))

    def generate_strings(self):
        return [self.generate_random_string() for _ in range(self.element_count)]

    def generate_strings_sorted(self, reversed=False):
        strings = self.generate_strings()
        strings.sort(reverse=reversed)
        return strings
```


Use the cell below to implement the requested **experimental framework** (be mindful of code readability and reusability).

```python
import timeit
import matplotlib.pyplot as plt

class ExperimentalFramework():
    '''
    A class to represent an experimental framework.

    ...

    Attributes
    ----------

    [to be defined as part of the coursework]

    Methods
    -------

    [to be defined as part of the coursework]

    '''

    #ADD YOUR CODE HERE

    def __init__(self):
        self.insert_results = {}
        self.search_results = {}
        pass

    def run_experiment(self, trees, data, checkpoints):
        for name,tree_class in trees.items():
            tree = tree_class()
            prev = 0
            for checkpoint in checkpoints:
                for x in data[prev:checkpoint]:
                    tree.insertElement(x)

                # Since we're taking an average over an interval of insertion,
                # it makes sense to scale the amount of samples we take based on
                # however many were already inserted. A sample of 50 when we're at 100 elements
                # inserted will be a lot noisier than a sample of 50 at 50000 elements
                batch_size = max(5, checkpoint // 10)

                # Insertion
                insert_batch = data[checkpoint:checkpoint+batch_size]
                times = []
                for x in insert_batch:
                    t = timeit.timeit(lambda e=x: tree.insertElement(e), number=1)
                    times.append(t)
                times.sort()

                if name not in self.insert_results:
                    self.insert_results[name] = {}
                self.insert_results[name][checkpoint] = times[len(times) // 2]

                # Searching
                search_batch = random.sample(data[:checkpoint], batch_size)
                times = timeit.repeat(lambda b=search_batch: [tree.searchElement(x) for x in b], number=1, repeat=5)
                times.sort()

                if name not in self.search_results:
                    self.search_results[name] = {}
                self.search_results[name][checkpoint] = times[len(times) // 2] / len(search_batch)

                prev = checkpoint + batch_size

    def plot_insert_times(self, title="Insert Performance"):
        plt.figure()
        for name, timings in self.insert_results.items():
            sizes = list(timings.keys())
            times = list(timings.values())
            plt.plot(sizes, times, label=name)
        plt.xlabel("Number of Elements")
        plt.ylabel("Average Insert Time (seconds)")
        plt.title(title)
        plt.legend()
        plt.show()

    def plot_search_times(self, title="Search Performance"):
        plt.figure()
        for name, timings in self.search_results.items():
            sizes = list(timings.keys())
            times = list(timings.values())
            plt.plot(sizes, times, label=name)
        plt.xlabel("Number of Elements")
        plt.ylabel("Average Search Time (seconds)")
        plt.title(title)
        plt.legend()
        plt.show()
```
